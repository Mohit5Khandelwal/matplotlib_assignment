Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data 
plot a scatter plot.

import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))

Note: Also add title, xlabel, and ylabel to the plot

A scatter plot is a type of plot that displays the relationship between two numeric variables. It represents the data as a collection of points on a two-dimensional coordinate system. Each point in the scatter plot represents the values of the two variables being plotted.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed( 3 )

x = 3 + np.random.normal( 0 , 2 , 50 )
y = 3 + np.random.normal( 0 , 2 , len( x ) )
color = np.random.rand( 50 )

plt.scatter( x , y , c = color , s = 100 , alpha=0.6)
plt.xlabel( "X-Axis" )
plt.ylabel( "Y-Axis" )
plt.title( "Scatter plot" )
```

The subplot() function in Matplotlib is used to create a grid of subplots within a single figure. It allows you to divide the figure into multiple smaller axes or subplots, each of which can contain a separate plot or visualization. 

```python
import numpy as np
```

```python
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 100, 200, 300, 400, 500])

plt.subplot( 2 , 2 , 1 )
plt.title( "Fig - 1")
plt.plot( x,y )

# Fig -2
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([50, 20, 40, 20, 60, 70])

plt.subplot( 2 , 2 , 2)
plt.title( "Fig - 2")
plt.plot( x,y )

# Fig -3
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([10, 20, 30, 40, 50, 60])

plt.subplot( 2 , 2 , 3)
plt.title( "Fig - 3")
plt.plot( x,y )

# Fig -4
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([200, 350, 250, 550, 450, 150])

plt.subplot( 2 , 2 , 4)
plt.title( "Fig - 4")
plt.plot( x,y )




```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot. 

import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])

A bar plot is a type of plot that represents categorical data using rectangular bars. It displays the relationship between a categorical variable and a numerical variable. The length or height of each bar corresponds to the value of the numerical variable associated with the specific category.

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])

plt.bar( company , profit )
plt.xlabel( "Company" )
plt.ylabel( "Profit" )
plt.title( "Bar Plot" )
```

```python
# Horizontal  Bar Plot 

plt.barh( company , profit )
plt.xlabel( "Company" )
plt.ylabel( "Profit" )
plt.title( "Horizontal Bar Plot" )
```

box1 = np.random.normal(100, 10, 200)

```python
box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)

plt.subplot( 1 , 2 , 1 )
plt.boxplot( box1 )
plt.show()

plt.subplot( 1 , 2 , 2 )
plt.boxplot( box2 )
plt.show()
```
